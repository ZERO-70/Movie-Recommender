```python
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Download the MovieLens dataset into the Colab environment if it doesn't exist
if not os.path.exists('data/ml-latest-small'):
    os.makedirs('data', exist_ok=True)
    # Using shell commands (!) to download and unzip the dataset quietly (-q)
    !curl -O https://files.grouplens.org/datasets/movielens/ml-latest-small.zip
    !unzip -q ml-latest-small.zip -d data/
    !rm ml-latest-small.zip
    print("Dataset downloaded and extracted successfully!")

# 2. Load the movies data into a Pandas DataFrame
movies_df = pd.read_csv('data/ml-latest-small/movies.csv')

# Display the first few rows to see the 'movieId', 'title', and 'genres' columns
print("\n--- First 5 rows of movies.csv ---")
display(movies_df.head())

# 3. Clean the genres column
# The genres are separated by '|' (e.g., "Action|Sci-Fi"). 
# We replace '|' with spaces so the vectorizer treats them as separate words.
movies_df['genres_cleaned'] = movies_df['genres'].str.replace('|', ' ')

# 4. Initialize and fit the TF-IDF Vectorizer
# This converts our text genres into a mathematical matrix of features.
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['genres_cleaned'])

# Check the shape of our new matrix
print(f"\nTF-IDF Matrix Shape: {tfidf_matrix.shape}")
print(f"We have {tfidf_matrix.shape[0]} movies, represented by {tfidf_matrix.shape[1]} unique genre features.")
```
